```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the Dataset
df = pd.read_csv(r'C:\Users\13kum\Downloads\titanic\train.csv')  # Make sure "titanic.csv" is in the same directory, or provide the full path

# 2. Initial Exploration
print("--- First 5 rows ---")
print(df.head())

print("\n--- DataFrame Info ---")
df.info()

print("\n--- Descriptive Statistics ---")
print(df.describe())

print("\n--- пассажир Class Value Counts ---") #Translates to "Passenger"
print(df['Pclass'].value_counts())

print("\n--- Sex Value Counts ---")
print(df['Sex'].value_counts())

print("\n--- Embarked Value Counts ---")
print(df['Embarked'].value_counts())

# 3. Univariate Analysis

# Numerical Variables
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), kde=True) # Drop NaN values for plotting
plt.title('Age Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Age'].dropna())
plt.title('Age Boxplot')

plt.show()

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Fare'].dropna(), kde=True)
plt.title('Fare Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Fare'].dropna())
plt.title('Fare Boxplot')
plt.show()

# Categorical Variables

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df, hue='Survived')
plt.title('Survival by Sex')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df, hue='Survived')
plt.title('Survival by Passenger Class')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', data=df, hue='Survived')
plt.title('Survival by Embarked Port')
plt.show()

# 4. Bivariate Analysis

# Numerical vs. Numerical
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs. Fare')
plt.show()

# Numerical vs. Categorical
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Passenger Class vs. Age')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Survived vs. Fare')
plt.show()

# Categorical vs. Categorical
# Using crosstab and stacked bar chart to visualize the relationship

# Survival vs. Sex
cross_tab_sex = pd.crosstab(df['Sex'], df['Survived'])
print("\n--- Survival vs. Sex Crosstab ---")
print(cross_tab_sex)

cross_tab_sex.plot(kind='bar', stacked=False)
plt.title('Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# Survival vs. Pclass
cross_tab_pclass = pd.crosstab(df['Pclass'], df['survived'])
print("\n--- Survival vs. Pclass Crosstab ---")
print(cross_tab_pclass)

cross_tab_pclass.plot(kind='bar', stacked=False)
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

# 5. Multivariate Analysis

# Pairplot (Numerical Variables)
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].dropna())
plt.title('Pairplot of Numerical Variables')
plt.show()

# Heatmap (Correlation Matrix)
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Observations and Insights (Example - You need to add more detailed observations)

#   * Age is slightly right-skewed.
#   * Fare is heavily right-skewed with many outliers.
#   * Survival rate is around 38%
#   * Females had a higher survival rate than males.
#   * Higher passenger classes had a higher survival rate.
#   * Age and Fare don't show a strong linear correlation.
#   * Pclass is negatively correlated with Survival.

# 6. Summary of Findings (You need to expand on this)

#   * The EDA reveals important relationships between variables like Sex, Pclass, and Survival.
#   * Further analysis could involve feature engineering (e.g., creating family size) and more advanced statistical testing.
```
